```python
new_df1_target0=base_data.loc[base_data["NoShow"]==0]
new_df1_target1=base_data.loc[base_data["NoShow"]==1]
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.figure(figsize=(10,6))
    ax = sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index,
        hue=hue,
        palette='bright'
    )
    plt.xticks(rotation=45)
    plt.title(title, fontsize=16, pad=20)
    plt.ylabel("Count")
    plt.show()
```

```python
uniplot(new_df1_target1,col='Hypertension',title='Distribution of Gender for NoShow Customers',hue='Gender')
```

```python
uniplot(new_df1_target1,col='Age_group',title='Distribution of Age for NoShow Customers',hue='Gender')
```

### Findings

1. Dataset Overview
The dataset contains 110,527 appointment records with 13 variables, including demographics (Gender, Age), medical conditions (Hypertension, Diabetes, Alcoholism, Handicap), socioâ€‘economic factors (Scholarship), communication (SMSReceived), and scheduling details (ScheduledDay, AppointmentDay, weekdays).

The target variable is NoShow, which indicates whether a patient attended their appointment.

2. Demographics
Gender: Both male and female patients are represented. From the preview, females appear more frequent in the dataset.

Age:

The average age is 37 years.

The youngest recorded age is -1 (likely a data entry error).

The oldest recorded age is 115.

25% of patients are under 18, 50% are under 37, and 75% are under 55.

3. Socioâ€‘Economic Factors
Scholarship: Only about 9.8% of patients have a scholarship (â‰ˆ10,800 patients).

Insight: The majority of patients do not benefit from this program.

4. Medical Conditions
Hypertension: ~19.7% of patients (â‰ˆ21,800) have hypertension.

Diabetes: ~7.2% of patients (â‰ˆ7,900) have diabetes.

Alcoholism: ~3% of patients (â‰ˆ3,300) are flagged for alcoholism.

Handicap: Most patients have no handicap; the maximum recorded value is 4, but the mean is only 0.02, showing very few with disabilities.

5. Communication
SMSReceived: About 32% of patients (â‰ˆ35,000) received an SMS reminder, while the majority did not.

Insight: This variable will be important to test whether reminders improve attendance.

6. Scheduling Patterns
ScheduledDay vs AppointmentDay:

Appointments were scheduled between Nov 2015 and June 2016.

Most appointments are concentrated in Aprilâ€“June 2016.

Weekdays: Both scheduled and appointment weekdays are captured numerically (0â€“6). This allows analysis of weekday patterns in attendance.

7. Target Variable (NoShow)
The bar chart shows the distribution of NoShow.

From the counts:

75,045 patients showed up.

35,482 patients did not show up.

This means about 32% of appointments were missed.

Insight: Noâ€‘Shows are a significant issue, with nearly 1 in 3 patients failing to attend.

Overall Insights

Age and chronic conditions (Hypertension, Diabetes) will likely be strong predictors of attendance.


SMS reminders and scholarship status are key behavioral/socioâ€‘economic factors to test.

The Noâ€‘Show rate of ~32% highlights the importance of predictive modeling and targeted interventions.
